```python
fig = go.Figure(go.Scatter(
    y=data["state"],
    x=data["no. of ratings"],
    text=data["location_name"],
    mode='markers',



))
# Tune marker appearance and layout
fig.update_traces(mode='markers', marker=dict(sizemode='area'
                                             , line_width=2))


fig.update_layout(
    title='attractions per state',
    xaxis=dict(
        title='user rating',
        gridcolor='white',
        type='log',
        gridwidth=2,
    ),
    yaxis=dict(
        title='states in india',
        gridcolor='white',
        gridwidth=2,
    )

)

fig.show()
```
