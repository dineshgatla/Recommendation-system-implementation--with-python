```python
import requests
from bs4 import BeautifulSoup as soup
```

```python
url = "https://www.holidify.com/country/india/places-to-visit.html"
html = requests.get(url)
html.status_code
```

```python
bsobj = soup(html.content, 'lxml')
```

```python
place1 = []
for name in bsobj.findAll('h3', {'class': 'card-heading'}):
    location_name = ''.join([i for i in name.text if not i.isdigit()]).strip().replace(".", "")
    place1.append(location_name)

place1
```

```python
ratings1 = []
for rating in bsobj.findAll('span', {'class': 'rating-badge'}):
    rating_value = rating.find('b').text
    ratings1.append(rating_value)
ratings1
```

```python
description1 = []
for name in bsobj.findAll('p', {'class': 'card-text'}):
  description1.append(name.text.strip())

description1
```

```python
best_time_to_visit1 = []
for name in bsobj.findAll('p', {'class': 'mb-3'}):
    text = name.text.strip()
    if text.startswith('Best Time:'):
        months = text.replace('Best Time:', '').strip()
        best_time_to_visit1.append(months)

best_time_to_visit1
```

```python
d1 = {'location_name': place1, 'ratings': ratings1, 'description': description1, 'best time to visit': best_time_to_visit1}
```

```python
import pandas as pd
```

```python
df_1 = pd.DataFrame.from_dict(d1)
df_1.head
```
